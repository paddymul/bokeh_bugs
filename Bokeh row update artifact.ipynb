```python
import ipywidgets as widgets
import numpy as np

from bokeh.layouts import column, row, widgetbox
from bokeh.models import CustomJS, ColumnDataSource
from bokeh.models.widgets import CheckboxGroup
from bokeh.models.renderers import GlyphRenderer
from bokeh.plotting import figure
from bokeh.io import output_notebook, push_notebook, show
from IPython.display import display
from bokeh.plotting.figure import Figure
import time

output_notebook()
```

```python
x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]

# create a new plot
s1 = figure(width=250, plot_height=250, title=None)
s1.circle(x, y0, size=10, color="navy", alpha=0.5)

# create another one
s2 = figure(width=250, height=250, title=None)
s2.triangle(x, y1, size=10, color="firebrick", alpha=0.5)


# create and another
s3 = figure(width=250, height=250, title=None)
s3.square(x, y2, size=10, color="olive", alpha=0.5)

# put the results in a column and show
#show(column(s1, s2, s3))
col_obj = column(row(s1,s2), row(s3))
```

```python
#this works properly with no artifacts
displayed_row = row(s1,s2)
handle = show(displayed_row, notebook_handle=True)
time.sleep(5)
del(displayed_row.children[0])
push_notebook(handle=handle)
```

```python
#this works properly with not artifacts
r1 = row(s1,s2)
displayed_obj = column(r1)
handle = show(displayed_obj, notebook_handle=True)
time.sleep(5)
del(r1.children[0])
push_notebook(handle=handle)
```

```python
#this messes up the rendering row2
r1 = row(s1,s2)
r2 = row(s3)
displayed_obj = column(r1, r2)

handle = show(displayed_obj, notebook_handle=True)
time.sleep(5)
del(r1.children[0])
push_notebook(handle=handle)
```
